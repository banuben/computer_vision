```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Basic Line Plot **

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

# **Line Styles and Colors**

```python
x = np.array([1,2,3,4])
y = np.array([3,8,1,10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.grid(True)
plt.show()
```

# Multiple Lines in One Graph with Numpy and Matplotlib

```python
x = np.array([1,2,3,4])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(x,y1, label='line 1', linestyle='dotted', color='red', marker='s')
plt.plot(x,y2, label='line 2', linestyle='dashed', color='green', marker='o')
plt.title("Multiple Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Chart with Numpy and **Matplotlib**

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# **Scatter Plot with Numpy and Matplotlib**

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y, color='red', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

# Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 5, 6, 7, 8, 9, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
# plt.grid(True)
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# Subplots(Multiple Graphs Together)

```python
x = np.array([1,2,3,4])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1, label='line 1', linestyle='dotted', color='red', marker='s')
plt.title("Subplot 1")
plt.grid(True)

plt.subplot(1,2,2)
plt.bar(x,y2, label='line 2', color='green')
plt.title("Subplot 2")

plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([3,8,1,10])

plt.plot(x, y, color='green')
plt.title("Save Figure Example")
plt.xlabel("x values")
plt.ylabel("y values")
plt.savefig('my_plot.png')
plt.show()
```
